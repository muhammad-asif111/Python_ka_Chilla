# 16- Spliting a String into multiple columns

## Making DataFrame

```python
import pandas as pd
df = pd.DataFrame({'name':['Muhammad Asif', 'Muhammad Ali', 'Irfan kazim', 'Usman Ghani'], 
                    'location': ['Bahawalpur, Pakistan', 'Multan,  Pakistan', 'Karachi, Pakistan', 'Lahore, Pakistan']})
df
```

## Spliting a columns into two columns 

```python
df.name.str.split(' ', expand=True)
```

## Adding those into two new columns

```python
df[["first_name", "last_name"]] = df.name.str.split(' ', expand=True)
df
```

## Expand Location

```python
df.location.str.split(', ', expand=True)
```

```python
df[["city", "country"]] = df.location.str.split(', ', expand=True)
df
```

## Refine Data Manipilation

```python
df = df[['first_name', 'last_name', 'city', 'country']]
df
```
