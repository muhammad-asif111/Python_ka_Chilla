# 16- Spliting a string into multiple Columns

```python
import pandas as pd
df = pd.DataFrame({'name': ['Abu Bakar', 'Umer Farooq', 'Muhammad Usman', 'Muhammad Ali'], 'location': ['Lahore, Pakistan', 'Karachi , Pakistan', 'Multan , Pakistan', 'London , Europe']})
df
```

```python
# Spliting into two Columns
df.name.str.split(' ', expand=True)
```

```python
# Adding those into new columns
df[["First Name", "Last Name"]] = df.name.str.split(' ', expand=True)
df
```

```python
df[["City", "Country"]] = df.location.str.split(', ' , expand=True)
df
```

```python
# Refine Data Manipolation 
df = df[['First Name', 'Last Name', 'City', 'Country']]
df
```
