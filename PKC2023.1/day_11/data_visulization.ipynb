```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
kashti = sns.load_dataset('titanic')
```

```python
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="survived", hue="sex", data=kashti)
```

```python
# Scatter Plot
plt.scatter(x="age", y="fare", data=kashti)
```

```python
sns.scatterplot(x="age", y="fare", hue="pclass", data=kashti)
```

```python
sns.barplot(x="sex", y="fare", hue="pclass", data=kashti)
```

```python
kashti.columns
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=10, figsize=(9,7),grid=False)
```

```python
sns.boxplot(data=kashti,x="sex", y="fare", hue="pclass")
```

```python
sns.boxenplot(data=kashti, x="sex", y="fare", hue="class")
```

```python
g= sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True, palette= {1: "seagreen", 0: "gray"})
g=g.map(plt.scatter, "fare", "age", edgecolor="g").add_legend();
```

```python
g= sns.FacetGrid(kashti, hue="survived", col="who", margin_titles=True, palette= {1: "seagreen", 0: "gray"})
g=g.map(plt.scatter, "fare", "age", edgecolor="g").add_legend();
```

```python
sns.stripplot(data=kashti, x="sex", y="fare", hue="class")
```

```python
corr = kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Heat Map of Kashit Data")
```

```python
sns.violinplot(data=kashti, x="sex", y="fare", hue="class")
```
